## Resolução

### Resolução analítica de (1.a), obtendo aceleração $a_i$

Partindo da expressão de MRUV, em que a aceleração $a$ é constante e $t$ é medido em relação a $t_0$ :

$$ x(t) = x_{0} + v_{0}\,(t - t_0) + \frac{a\,(t - t_0)^2}{2}$$

Como $v_{0}$ é nulo pois o carrinho estava em repouso inicialmente, e agrupando $x$ :

$$[ x(t) - x_{0}] =  \frac{a\,(t - t_0)^2}{2}$$

Explicitando a aceleração $a$ :

$$ a = \frac{2\,[ x(t) - x_{0}]}{(t - t_0)^2}$$

Usando indice $i$ :

$$ a_i = \frac{2\,( x_i - x_{0})}{(t_i - t_0)^2}$$

Se o cálculo de $a_i$ com propagação de incertezas fosse feito manualmente, teria 5 cálculos manuais para cada $i$ : 2 subtrações, 1 potência, 1 divisão, 1 produto. O que gastaria várias linhas no papel para cada $i$. 

### Uso do [módulo Uncertainties](https://pythonhosted.org/uncertainties/) para definir e propagar incertezas

O cálculo da aceleração $a_i$, etc, com propagação de incertezas, deve usar o módulo Python Uncertainties pois foi assim pedido no enunciado do problema.

A vantagem do  módulo Python Uncertainties sobre cálculo manual de propagação de incertezas é maior eficiência : basta escrever a expressão matemática dos termos com incertezas que o resultado final é obtido com propagação automática das incertezas.

Isso poupa muito tempo e evita erros humanos nos cálculos, ainda mais quando o número de medidas é grande.

#### Definições de números com incertezas via módulo Uncertainties

Vide documentação ["Creating numbers with uncertainties - User Guide"](https://pythonhosted.org/uncertainties/user_guide.html#creating-numbers-with-uncertainties), basicamente se deve usar a função `ufloat(valor exato, erro)` para definir números com incertezas.

Vamos primeiramente definir $x_0$, $t_0$, $x_1$, $t_1$ com os valores dados no enunciado :

```python
x0 = ufloat(0.000, 0.005); x0
```

```python
t0 = ufloat(0.000, 0.005); t0
```

```python
x1 = ufloat(0.021, 0.005); x1
```

```python
t1 = ufloat(0.120, 0.005); t1
```

Um número com incerteza definido via `ufloat` pode ter suas partes (exata/nominal e incerteza/erro/desvio padrão) acessíveis em separado, vide documentação ["Access to the uncertainty and to the nominal value - User Guide"](https://pythonhosted.org/uncertainties/user_guide.html#access-to-the-uncertainty-and-to-the-nominal-value).

`.n` e `.nominal_value` são sinônimos para parte exata/nominal do valor com incerteza :

```python
[x1.n, x1.nominal_value]
```

`.s` e `.std_dev` são sinônimos para parte de incerteza/erro/desvio padrão do valor com incerteza :

```python
[x1.s, x1.std_dev]
```

#### Cálculo de $a_1$ com propagação de incertezas usando módulo Uncertainties

O cálculo de propagação de incertezas usando módulo Uncertainties é trivial, basta uma linha e o resultado final é obtido, com 15-16 algarismos.

Mas note que não é feita análise de número correto de algarismos significativos por parte do módulo Uncertainties, ao exibir o valor da variável com incerteza. 

```python
a1 = 2*(x1 - x0)/((t1 - t0)**2); a1
```

Porém usando saída formatada via `print()` (vide seção ["Printing - User Guide"](https://pythonhosted.org/uncertainties/user_guide.html#printing)), tem-se controle do número de algarismos significativos exibidos no valor exato/nominal e na incerteza/erro/desvio, via módulo Uncertainties :

```python
print(a1)
```

O que confere com o resultado com número correto de algarismos significativos, $a_1 = (2,9 \pm 1,0)\,m/s^2$, ao se considerar que incerteza deve ter no máximo 2 algarismos significativos, compondo uma dezena 'baixa' (convenção de cada um). Vide [material do Particle Data Group, seção "5.3. Rounding", pág. 13](https://pdg.lbl.gov/2010/reviews/rpp2010-rev-rpp-intro.pdf).

Embora os resultados parciais não sejam exibidos para cada operação matemática, pode-se calcular uma operação matemática por vez, por exemplo :

```python
(x1 - x0)
```

```python
2*(x1 - x0)
```

```python
(t1 - t0)
```

```python
(t1 - t0)**2
```

Via módulo Uncertainties, cada operação matemática com incertezas não tem os cálculos passo-a-passo mostrados.

#### 1.a) Cálculo de $a_i$ com propagação de incertezas usando módulo Uncertainties e arrays (de NumPy)

<center><img src="https://raw.githubusercontent.com/numpy/numpy/master/branding/logo/primary/numpylogo.png" alt="Logo de NumPy" width="600"/></center>

Um subconjunto de NumPy suporta definições com incertezas e propagação de incertezas, vide documentação  ["Uncertainties in arrays"](https://pythonhosted.org/uncertainties/numpy_guide.html).

Definindo as medições de posição $X = x \pm \Delta x$ e tempo $T = t \pm \Delta t$ da tabela 1 :

```python
x = unumpy.uarray([0.021, 0.072, 0.172, 0.293, 0.464], [0.005, 0.006, 0.008, 0.009, 0.010]); x
```

```python
t = unumpy.uarray([0.120, 0.240, 0.360, 0.480, 0.600], [0.005, 0.005, 0.005, 0.005, 0.005]); t
```

Saída formatada via `print()` não exibe os elementos do array com número correto de algarismos significativos :

```python
print("x +/- dx = ", x); print("t +/- dt = ", t)
```

Mas `print()` de cada valor do array exibe com número correto de algarismos significativos :

```python
print("x[4] = ", x[4]); print("t[4] = ", t[4])
```

```python
for xi in x:
    print(xi)
```

```python
for ti in t:
    print(ti)
```

Para extrair a parte exata/nominal e a parte e incerteza/erro/desvio padrão do valor com incerteza, vide [Uncertainties and nominal values - Uncertainties in arrays](https://pythonhosted.org/uncertainties/numpy_guide.html#uncertainties-and-nominal-values) :

```python
unumpy.nominal_values(x)
```

```python
unumpy.std_devs(x)
```

```python
unumpy.nominal_values(t)
```

```python
unumpy.std_devs(t)
```

Cálculos com incertezas de vários valores podem ser feitos sem precisar usar loop, graças a NumPy com Uncertainties, aqui todos valores $x_i$ menos $x_0$ : 

```python
x - x0
```

Tal que os valores de $a_i$ são calculados diretamente : 

```python
a = 2*(x - x0)/((t - t0)**2); a
```

Algo parecido com tabela usando `print()` em Python não é muito elegante : 

```python
print("i         t[i]             x[i]           a[i]")
for i in range(len(x)):
    print(i+1, "  ", t[i], "  ", x[i], "  ", a[i])
```
